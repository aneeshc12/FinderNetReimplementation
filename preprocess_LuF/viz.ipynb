```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os, sys
root_dir = '/scratch/aneesh.chavan/LuF_DEMs/inner'
files = os.listdir(root_dir)
```

```python
%matplotlib inline

fig, axes = plt.subplots(1,10,figsize=(5,30))
for i in range(10):
    dem = np.load(os.path.join(root_dir, files[i]))
    axes[i].imshow(dem)
fig.show()
```
